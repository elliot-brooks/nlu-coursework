```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data/inputs and configuration
models = ['Accuracy', 'Precision', 'Recall', 'F-score']
metrics = ['Baseline SVM', 'BERT Embedding CNN', 'Baseline Bi-LSTM', 'BERT Embedding Bi-LSTM']
scores = {
    'Accuracy': [0.53, 0.60, 0.50, 0.54],
    'Precision': [0.53, 0.61, 0.50, 0.54],
    'Recall': [0.53, 0.59, 0.50, 0.54],
    'F-score': [0.53, 0.58, 0.50, 0.54]
}

colors = ['#bfbfbf', '#b797cf', '#3a3a3a', '#9464b8']
patterns = ['//', '', '//', '']
```

```python
# Create a bar chart
bar_width = 0.2
index = np.arange(len(models))
fig, ax = plt.subplots()
for i, metric in enumerate(metrics):
    scores_for_metric = [scores[model][i] for model in models]
    ax.bar(index + i * bar_width, scores_for_metric, bar_width, label=metric, color=colors[i], hatch=patterns[i])

ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_ylim([0, 1])
ax.set_title('Performance of RNN Bi-LSTM and CNN with BERT Embeddings compared to baseline models')
ax.set_xticks(index + bar_width * 1.5)
ax.set_xticklabels(models)
ax.legend()

plt.tight_layout()
plt.show()
```
